```
%%bash 
mkdir wavs
for file in "./mp3/"*.mp3; do
	ffmpeg -i "$file" -ar 16000 -ac 1 "./wavs/""`basename "$file" .mp3`.wav"
done 
```

```
%%bash
ls wavs/*.wav | sort -R --random-source=/dev/urandom > wav_files.txt
cat wav_files.txt | tail -n+10 > train_files.txt
cat wav_files.txt | head -n10 > test_files.txt

echo "\n\n ===== wavs ===== \n\n"
ls wavs

echo "===== train ====="
cat train_files.txt

echo "===== test ====="
cat test_files.txt
```
